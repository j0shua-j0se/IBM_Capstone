```python
import folium
from folium.features import DivIcon
import pandas as pd
from js import fetch
import io

# URL to the dataset
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'

# Fetch the dataset
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Select relevant sub-columns
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]

# Start location is NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Add circles and markers for each launch site
for _, site in launch_sites_df.iterrows():
    coordinate = [site['Lat'], site['Long']]
    folium.Circle(
        coordinate, radius=1000, color='#000000', fill=True
    ).add_child(folium.Popup(site['Launch Site'])).add_to(site_map)
    
    folium.Marker(
        coordinate,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % site['Launch Site'],
        )
    ).add_to(site_map)

# Display the map
site_map
```

```python
# Task 2: Mark the success/failed launches for each site on the map
import folium
from folium.features import DivIcon
from folium.plugins import MarkerCluster

# Fetch the dataset
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Initialize the map (use previously created map if it exists)
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Initialize MarkerCluster
marker_cluster = MarkerCluster().add_to(site_map)

# Define marker colors based on launch success
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Add markers for each launch record
for _, record in spacex_df.iterrows():
    coordinate = [record['Lat'], record['Long']]
    folium.Marker(
        coordinate,
        icon=folium.Icon(color='white', icon_color=record['marker_color']),
        popup=f"Site: {record['Launch Site']}\nSuccess: {record['class']}"
    ).add_to(marker_cluster)

# Display the updated map
site_map
```

```python
import folium
from folium.plugins import MarkerCluster

# Fetch the dataset
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Initialize the map (use previously created map if it exists)
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Initialize MarkerCluster
marker_cluster = MarkerCluster().add_to(site_map)

# Define marker colors based on launch success
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Add markers for each launch record
for _, record in spacex_df.iterrows():
    coordinate = [record['Lat'], record['Long']]
    marker = folium.Marker(
        location=coordinate,
        icon=folium.Icon(color='white', icon_color=record['marker_color']),
        popup=f"Site: {record['Launch Site']}\nSuccess: {record['class']}"
    )
    marker_cluster.add_child(marker)

# Add marker_cluster to the map
site_map.add_child(marker_cluster)

# Display the updated map
site_map
```

```python
import folium
from folium.features import DivIcon

# Coordinates for the launch site (replace with actual values)
launch_site_coords = [28.56367, -80.57163]  # Example coordinates for launch site

# Example coordinates for other points of interest (replace with actual values)
city_coords = [28.56000, -80.57000]  # Coordinates for the closest city
railway_coords = [28.56500, -80.57500]  # Coordinates for the closest railway
highway_coords = [28.57000, -80.58000]  # Coordinates for the closest highway

# Create markers for each point of interest with distance information
def add_marker_with_distance(map_obj, point_coords, label):
    distance = calculate_distance(launch_site_coords[0], launch_site_coords[1], point_coords[0], point_coords[1])
    folium.Marker(
        point_coords,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12; color:#d35400;"><b>{distance:.2f} KM</b></div>',
        ),
        popup=f"Distance to Launch Site: {distance:.2f} KM"
    ).add_to(map_obj)

# Add markers
add_marker_with_distance(site_map, city_coords, "City")
add_marker_with_distance(site_map, railway_coords, "Railway")
add_marker_with_distance(site_map, highway_coords, "Highway")

# Draw lines between the launch site and each point of interest
def add_line_between_points(map_obj, start_coords, end_coords):
    folium.PolyLine(
        locations=[start_coords, end_coords],
        color='blue',
        weight=2
    ).add_to(map_obj)

# Add lines
add_line_between_points(site_map, launch_site_coords, city_coords)
add_line_between_points(site_map, launch_site_coords, railway_coords)
add_line_between_points(site_map, launch_site_coords, highway_coords)

# Display the updated map
site_map
```
